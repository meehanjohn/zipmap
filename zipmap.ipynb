```python
from geopy import distance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zip_df = pd.read_csv("zip_lat_long.csv", sep=';')
```

```python
zip_df.head()
```

```python
zip_df = zip_df[['Zip', 'State', 'Latitude', 'Longitude']]
```

```python
zip_df.Zip = zip_df.Zip.astype(str)
```

```python
zip_df.Zip = zip_df.Zip.apply(lambda x: '0'*(5-len(x))+x)
```

```python
piney_flats = {'Zip':'37686','Latitude':None,'Longitude':None}
roanoke = {'Zip':'24012','Latitude':None,'Longitude':None}
fontana = {'Zip':'92337','Latitude':None,'Longitude':None}
greeneville = {'Zip':'37745','Latitude':None,'Longitude':None}
lancaster = {'Zip':'17601','Latitude':None,'Longitude':None}
bethlehem = {'Zip':'18015','Latitude':None,'Longitude':None}
suffolk = {'Zip':'23434','Latitude':None,'Longitude':None}
florence = {'Zip':'29501','Latitude':None,'Longitude':None}
rocky_mount = {'Zip':'27801','Latitude':None,'Longitude':None}
ontario = {'Zip':'91764','Latitude':None,'Longitude':None}
```

```python
piney_flats['Latitude'] = zip_df[zip_df.Zip == piney_flats['Zip']].Latitude.item()
piney_flats['Longitude'] = zip_df[zip_df.Zip == piney_flats['Zip']].Longitude.item()

roanoke['Latitude'] = zip_df[zip_df.Zip == roanoke['Zip']].Latitude.item()
roanoke['Longitude'] = zip_df[zip_df.Zip == roanoke['Zip']].Longitude.item()

fontana['Latitude'] = zip_df[zip_df.Zip == fontana['Zip']].Latitude.item()
fontana['Longitude'] = zip_df[zip_df.Zip == fontana['Zip']].Longitude.item()

greeneville['Latitude'] = zip_df[zip_df.Zip == greeneville['Zip']].Latitude.item()
greeneville['Longitude'] = zip_df[zip_df.Zip == greeneville['Zip']].Longitude.item()

lancaster['Latitude'] = zip_df[zip_df.Zip == lancaster['Zip']].Latitude.item()
lancaster['Longitude'] = zip_df[zip_df.Zip == lancaster['Zip']].Longitude.item()

bethlehem['Latitude'] = zip_df[zip_df.Zip == bethlehem['Zip']].Latitude.item()
bethlehem['Longitude'] = zip_df[zip_df.Zip == bethlehem['Zip']].Longitude.item()

suffolk['Latitude'] = zip_df[zip_df.Zip == suffolk['Zip']].Latitude.item()
suffolk['Longitude'] = zip_df[zip_df.Zip == suffolk['Zip']].Longitude.item()

florence['Latitude'] = zip_df[zip_df.Zip == florence['Zip']].Latitude.item()
florence['Longitude'] = zip_df[zip_df.Zip == florence['Zip']].Longitude.item()

rocky_mount['Latitude'] = zip_df[zip_df.Zip == rocky_mount['Zip']].Latitude.item()
rocky_mount['Longitude'] = zip_df[zip_df.Zip == rocky_mount['Zip']].Longitude.item()

ontario['Latitude'] = zip_df[zip_df.Zip == ontario['Zip']].Latitude.item()
ontario['Longitude'] = zip_df[zip_df.Zip == ontario['Zip']].Longitude.item()
```

```python
def get_dist(lat, long, fc):
    beg = (fc['Latitude'], fc['Longitude'])
    end = (lat, long)
    dist = int(distance.distance(beg, end).mi)
    return(dist)
```

```python
zip_df['piney_flats_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, piney_flats), axis=1)
zip_df['roanoke_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, roanoke), axis=1)
zip_df['fontana_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, fontana), axis=1)
zip_df['greeneville_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, greeneville), axis=1)
```

```python
zip_df['lancaster_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, lancaster), axis=1)
zip_df['bethlehem_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, bethlehem), axis=1)
zip_df['suffolk_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, suffolk), axis=1)
zip_df['florence_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, florence), axis=1)
zip_df['rocky_mount_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, rocky_mount), axis=1)
zip_df['ontario_dist'] = zip_df[['Latitude', 'Longitude']].apply(lambda x: get_dist(*x, ontario), axis=1)
```

```python
zip_df.to_csv('fc_zip_dist.csv', index=False)
```
