```python
zone_df = zone_df.dropna()
```

```python
zone_df.loc[zone_df['rocky_mount_zone'] == 2, 'rocky_mount_dist'].values[:, np.newaxis]
```

```python
fc_list = ['florence', 'suffolk', 'ontario', 'lancaster', 'rocky_mount']
```

```python
X_plot = np.linspace(0, 2500, 1000)[:, np.newaxis]
cmap = matplotlib.cm.get_cmap('plasma')
norm = matplotlib.colors.Normalize(vmin=1, vmax=10)

fig, ax = plt.subplots(nrows=5, ncols=1, sharex=True, sharey=True, figsize=(10,18))

for i, fc in enumerate(fc_list):
    zone = fc+"_zone"
    dist = fc+"_dist"
    legend_text = []
    kde_funcs = []
    lower_bounds = []
    upper_bounds = []
    
    for j in range(2,9):

        kde_skl = KernelDensity(kernel='gaussian', bandwidth=10)
        
        kde_skl.fit(
            zone_df.loc[zone_df[zone] == j, dist].values[:, np.newaxis]
        )
        log_dens = kde_skl.score_samples(X_plot)
        
        ax[i].plot(
            X_plot[:, 0], 
            np.exp(log_dens), 
            c=cmap(norm(j)), 
            lw=2,      
            linestyle='-'
        )
        ax[i].hist(
            zone_df.loc[zone_df[zone] == j, dist].values, 
            color=cmap(norm(j)), 
            density=True, 
            alpha=0.5
        )
        
        ax[i].set_title("Zip Code Distance by Zone - {0}".format(fc))
        
        legend_text.append("Zone {0}".format(j))
        kde_funcs.append(kde_skl)
        
        if j>2:            
            fun1 = np.exp(kde_funcs[j-3].score_samples(X_plot))
            fun2 = np.exp(kde_funcs[j-2].score_samples(X_plot))

            g = fun1-fun2
            upper_bound = max(np.argwhere(g < 0.002*max(g)))[0]
            lower_bound = min(np.argwhere(g > 0.002*max(g)))[0]            
            
            idx = np.argwhere(np.diff(np.sign(g))).flatten()

            if j<6:
                result = X_plot[idx[0]]
            else:
                result = X_plot[idx[1]]
            
            ax[i].axvline(result, color=cmap(norm(j)), label='_nolegend_')
            
            ax[i].text(
                result+15,
                0.0125,
                str(round(result.item(),0)),
                rotation=90
            )
    
    ax[i].legend(legend_text, loc=1)
```
